# Exploratory Data Analysis (EDA)
## Patient Churn & Marketing Conversion Analytics

This notebook explores patterns, behaviors, and key drivers influencing patient churn and marketing conversion using interactive visualizations.

The objective of this analysis is to identify key drivers of patient churn and marketing conversion, enabling data-driven strategies for improving retention and optimizing marketing performance.

```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns

# Settings
sns.set(style="whitegrid")
```

## Data Cleaning and Preparation

In this section, we handle missing values, remove duplicates, and ensure data consistency before analysis.

```python
# Load datasets
churn_df = pd.read_csv('patient_churn_main.csv')
marketing_df = pd.read_csv('patient_conversion_marketing.csv')
validation_df = pd.read_csv('patient_churn_validation.csv')

# Preview
churn_df.head()
```

```python
# Shape
print("Churn Dataset Shape:", churn_df.shape)

# Info
churn_df.info()

# Summary stats
churn_df.describe()

# Missing values
churn_df.isnull().sum()
```

```python
# Remove duplicates
churn_df.drop_duplicates(inplace=True)

# Handle missing values
churn_df.fillna(method='ffill', inplace=True)

# Convert date column
churn_df['Last_Interaction_Date'] = pd.to_datetime(churn_df['Last_Interaction_Date'], errors='coerce')

# Check again
churn_df.isnull().sum()
```

# Feature Engineering

## Churn Risk Score

```python
# Create a churn risk score
churn_df['Risk_Score'] = (
    (10 - churn_df['Overall_Satisfaction']) +
    churn_df['Missed_Appointments'] +
    (12 - churn_df['Tenure_Months'] / 2)
)
```

```python
# Segment into risk categories
churn_df['Risk_Level'] = pd.cut(
    churn_df['Risk_Score'],
    bins=3,
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

churn_df[['Risk_Score', 'Risk_Level']].head()
```

## Risk Segmentation

```python
fig = px.histogram(
    churn_df,
    x='Risk_Level',
    color='Churned',
    barmode='group',
    title='Churn Risk Segmentation'
)
fig.show()
```

High-risk patients are clearly separated based on behavioral and satisfaction features, demonstrating the effectiveness of the engineered risk score.

## Age Group

```python
churn_df['Age_Group'] = pd.cut(
    churn_df['Age'],
    bins=[0, 30, 50, 70, 100],
    labels=['Young', 'Middle', 'Senior', 'Elder']
)

fig = px.histogram(
    churn_df,
    x='Age_Group',
    color='Churned',
    title='Churn by Age Group'
)
fig.show()
```

Different age groups exhibit varying churn patterns, enabling targeted retention strategies for specific demographics.

# Behavioural Pattern / Marketing Analysis

## Cost Efficiency

```python
marketing_df['Cost_Per_Conversion'] = marketing_df['AdSpend'] / (marketing_df['ConversionRate'] + 0.01)

fig = px.box(
    marketing_df,
    x='CampaignChannel',
    y='Cost_Per_Conversion',
    title='Cost Efficiency by Channel'
)
fig.show()
```

## Engagement Funnel

```python
fig = px.scatter(
    marketing_df,
    x='WebsiteVisits',
    y='ConversionRate',
    color='CampaignChannel',
    size='AdSpend',
    hover_data=['TimeOnSite'],
    title='Engagement vs Conversion Funnel'
)
fig.show()
```

Higher website visits and engagement levels lead to increased conversion rates, confirming that deeper interaction improves marketing effectiveness.

## Diminishing Returns

```python
fig = px.scatter(
    marketing_df,
    x='AdSpend',
    y='ConversionRate',
    trendline='ols',
    title='AdSpend vs Conversion Rate (Diminishing Returns)'
)
fig.show()
```

While increased ad spend improves conversion, the trendline indicates diminishing returns, suggesting that optimization is more important than simply increasing budget.

# Interaction Plot

```python
fig = px.scatter(
    churn_df,
    x='Missed_Appointments',
    y='Overall_Satisfaction',
    color='Churned',
    size='Tenure_Months',
    title='Behavior + Satisfaction + Tenure Interaction'
)
fig.show()
```

Patients with high missed appointments and low satisfaction, especially with shorter tenure, are most likely to churn. This highlights a multi-factor churn pattern.


## Final Business Insights

- Patient churn is primarily driven by satisfaction, engagement, and tenure  
- Early-stage patient experience is critical in preventing churn  
- Behavioral signals such as missed appointments act as early warning indicators  
- Digital engagement significantly improves retention outcomes  
- Marketing effectiveness is driven by user interaction depth rather than just ad spend  
- Increasing ad spend alone is insufficient due to diminishing returns
